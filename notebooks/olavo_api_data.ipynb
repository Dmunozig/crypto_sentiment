```python
import pandas as pd
import requests
dict = requests.get("http://api-dev.augmento.ai/v0.1/topics").json()
df = pd.DataFrame.from_dict(dict,orient='index',columns=["Themes"])
list_of_themes = ['Pessimistic/Doubtful', 'Selling', 'Angry', 'Bottom', 'Adoption', 'Fearful/Concerned', 'Flippening', 'Euphoric/Excited', 'Hodling', 'Bearish', 'Going_short', 'Uncertain', 'Risk', 'Ban', 'Cheap', 'Progress', 'Bullish', 'Happy', 'Bubble', 'Hopeful', 'Bad_news', 'Pump_and_dump', 'Sad', 'Panicking', 'Regulation/Politics', 'Dip','FOMO_theme', 'Good_news', 'Problems_and_issues', 'Recovery', 'Mistrustful', 'Positive', 'Long_term_investing', 'Correction', 'Buying', 'Warning', 'Annoyed/Frustrated', 'Scam/Fraud', 'Optimistic', 'Negative']

list_of_indexes = []
for themes in list_of_themes:
    list_of_indexes.append(int(df[df["Themes"]==themes].index[0]))
list_of_indexes   
```

```python
import requests

url = "http://api-dev.augmento.ai/v0.1/events/aggregated"

params = {
  "source" : "twitter",
  "coin" : "bitcoin",
  "bin_size" : "24H",
  "count_ptr" : 120,
  "start_ptr" : 30,
  "start_datetime" : "2018-02-01T00:00:00Z",
  "end_datetime" : "2020-04-30T00:00:00Z",
}

r = requests.request("GET", url, params=params)

list_of_dict = r.json()
```
