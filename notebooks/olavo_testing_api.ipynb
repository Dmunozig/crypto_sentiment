```python
from crypto_project.data import price_and_indexes
from crypto_project.model import fit_model,future_dataframe,model_predict, train_test_split
from fbprophet import Prophet
```

```python
%load_ext autoreload
%autoreload 2
```

```python
dataframe = price_and_indexes(local=True)
dataframe
```

```python
merged_model = Prophet(interval_width=0.95, weekly_seasonality=True, changepoint_prior_scale=2)

merged_model.add_regressor('BTC_score')
merged_model.add_regressor('twitter_score')
merged_model.add_regressor('reddit_score')
merged_model.add_regressor("Fear&Greed")

merged_model.fit(dataframe)
```

```python
augmento_horizon = 1
merged_model_future = merged_model.make_future_dataframe(augmento_horizon)

merged_model_future["BTC_score"] = dataframe["BTC_score"] 
merged_model_future["twitter_score"] = dataframe["twitter_score"] 
merged_model_future["reddit_score"] = dataframe["reddit_score"] 
merged_model_future["Fear&Greed"] = dataframe["Fear&Greed"] 
merged_model_future
```

```python
merged_model_forecast = merged_model.predict(merged_model_future)
```
